```python
%matplotlib notebook
from numpy import genfromtxt
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import time

## Full history
my_data=genfromtxt('position_full_history.csv',delimiter=',')
data_cutoff=0 # to cutoff bad data at the beginning
positions=np.array(my_data[data_cutoff:,1:])
timestamps=np.array(my_data[data_cutoff:,0])
dates=[dt.datetime.fromtimestamp(ts) for ts in timestamps]
#samplerate='1S'
df=pd.DataFrame({'timestamps': timestamps,'channel 0': positions[:,0],'channel 1': positions[:,1],'channel 2': positions[:,2],'channel 3': positions[:,3]
                },index=dates)

#data_to_plot1=df.resample(samplerate).mean()

## Only the final positions
my_data2=genfromtxt('position_final_history.csv',delimiter=',')
data_cutoff=0 # to cutoff bad data at the beginning
positions2=np.array(my_data2[data_cutoff:,1:])
timestamps2=np.array(my_data2[data_cutoff:,0])
dates2=[dt.datetime.fromtimestamp(ts) for ts in timestamps2]
df2=pd.DataFrame({'timestamps': timestamps2,
    'channel 0': positions2[:,0],
                             'channel 1': positions2[:,1],
                             'channel 2': positions2[:,2],
                             'channel 3': positions2[:,3]
                },index=dates2)

#data_to_plot2=df2.resample(samplerate).mean()
```

```python
figW=8
figH=6
fig, ax = plt.subplots()
fig.set_size_inches(figW,figH)

#current_time=dt.datetime.fromtimestamp(time.time())
draw_starttime=dt.datetime(2018,6,28,17)

df.plot(y=['channel 0','channel 1','channel 2','channel 3'],marker='.',lw=0,markersize=5,ax=ax)
ax.set_ylabel('Position(Degree)')
ax.grid()
ax.legend()
ax.set_xlim(draw_starttime,)
plt.show()

fig, ax = plt.subplots()
fig.set_size_inches(figW,figH)
df2.plot(y=['channel 0','channel 1','channel 2','channel 3'],marker='.',lw=0,markersize=5,ax=ax)
ax.set_ylabel('Position(Degree)')
ax.grid()
ax.legend()
ax.set_xlim(draw_starttime,)
plt.show()
```
